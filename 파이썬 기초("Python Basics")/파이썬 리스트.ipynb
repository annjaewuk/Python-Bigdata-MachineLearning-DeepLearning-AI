```python
name = list()
score = list()
student = 1

def getName() :
  
  while True:
    data = input("학생 이름: ");
    if data.isalpha() :
      return data
      break
    else :
      print("학생 이름은 문자여야 합니다.")

def getScore() :
  
  while True:
    data = input("학생 점수: ");
    if data.isdigit() :
      return int(data)
      break
    else :
      print("학생 점수는 정수여야 합니다.")


while True:
  name.append(getName())
  score.append(getScore())
  if len(name) == student:
    break

print(name, score)  
```
