```python
def map_count_words(sentence):
    words = sentence.strip().split()
    return [(word, 1) for word in words]

def reduce_count_words(word_counts):
    word_counts_dict = {}
    for word, count in word_counts:
        if word in word_counts_dict:
            word_counts_dict[word] += count
        else:
            word_counts_dict[word] = count
    return [(word, word_counts_dict[word]) for word in word_counts_dict]

if __name__ == '__main__':
    # 입력 데이터
    input_data = [
        "Hello world",
        "My name is John",
        "This is a test",
        "Hello again"
    ]

    # 맵 함수 수행
    mapped_data = []
    for sentence in input_data:
        mapped_data.extend(map_count_words(sentence))

    # 리듀스 함수 수행
    reduced_data = {}
    for word, count in reduce_count_words(mapped_data):
        if word in reduced_data:
            reduced_data[word] += count
        else:
            reduced_data[word] = count

    # 결과 출력
    for word, count in reduced_data.items():
        print(f"{word}: {count}")
```
