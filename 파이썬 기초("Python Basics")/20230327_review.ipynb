```python
#스택 예제 (역순으로 출력)

def reverse_string(text):
    stack = []
    for char in text:
        stack.append(char)

    reversed_text = ''
    while len(stack) != 0:
        reversed_text += stack.pop()

    return reversed_text

# 사용 예시
text = input('텍스트를 입력하세요: ')
reversed_text = reverse_string(text)
print('입력한 텍스트의 역순:', reversed_text)
```

```python
def print_string(text):
    queue = []
    for char in text:
        queue.append(char)

    printed_text = ''
    while len(queue) != 0:
        printed_text += queue.pop(0)

    return printed_text

# 사용 예시
text = input('텍스트를 입력하세요: ')
printed_text = print_string(text)
print('입력한 텍스트:', printed_text)
```

```python
class Customer:
    def __init__(self, name):
        self.name = name

class BankQueue:
    def __init__(self):
        self.queue = []

    def add_customer(self, customer):
        self.queue.append(customer)

    def serve_customer(self):
        if len(self.queue) == 0:
            print('대기 중인 고객이 없습니다.')
        else:
            customer = self.queue.pop(0)
            print(customer.name, '님을 상담 중입니다.')

# 사용 예시
bank_queue = BankQueue()

bank_queue.add_customer(Customer('김철수'))
bank_queue.add_customer(Customer('박영희'))
bank_queue.add_customer(Customer('최민식'))

bank_queue.serve_customer() 
bank_queue.serve_customer() 
bank_queue.serve_customer() 
bank_queue.serve_customer()
bank_queue.serve_customer()
```
