```python
class Animal:
    def __init__(self, species, name, age):
        self.species = species
        self.name = name
        self.age = age

    def say_hello(self):
        print(f"{self.name} is a {self.species} and is {self.age} years old.")

class Dog(Animal):
    def bark(self):
        return f"{self.name}이(가) 짖습니다.. 멍멍!"

    def user_input(self):
        while True:
            user_key = input("Press 'b' key to make the dog bark, 'q' to quit: ")
            if user_key == 'b':
                print(self.bark())
            elif user_key == 'q':
                break
            else:
                print("Invalid input. Please try again.")

class Cat(Animal):
    def meow(self):
        return f"{self.name}이(가) 소리를냅니다.. 야옹!"

    def user_input(self):
        while True:
            user_key = input("Press 'm' key to make the cat meow, 'q' to quit: ")
            if user_key == 'm':
                print(self.meow())
            elif user_key == 'q':
                break
            else:
                print("Invalid input. Please try again.")

def validate_int_input(input_str):
    while True:
        user_input = input(input_str)
        try:
            user_input = int(user_input)
            return user_input
        except ValueError:
            print("Invalid input. Please enter an integer.")

def validate_animal_type(animal_type):
    while animal_type != 'dog' and animal_type != 'cat':
        print("Invalid animal type. Please enter 'dog' or 'cat'.")
        animal_type = input("Enter the animal type: ")
    return animal_type

def main():
    animals = []

    while True:
        animal_type = input("Enter the animal type (dog/cat) or 'q' to quit: ")
        if animal_type == 'q':
            break
        animal_type = validate_animal_type(animal_type)

        name = input("Enter the animal name: ")
        age = validate_int_input("Enter the animal age: ")
        
        if animal_type == 'dog':
            my_animal = Dog(animal_type, name, age)
            my_animal.user_input()
            animals.append(my_animal)
        elif animal_type == 'cat':
            my_animal = Cat(animal_type, name, age)
            my_animal.user_input()
            animals.append(my_animal)

    if animals:
        print("\nAll animals:\n")
        for animal in animals:
            animal.say_hello()
    else:
        print("No animals entered.")

if __name__ == '__main__':
    main()
```
