```python
#@title 예제 풀이

i = 0
totalList = []
keyList = []
valueList = []

while True :
  name = input("이름 :")
  integer = input("정수 :")
  dic = {name: int(integer)}
  totalList.append(dic)
  i = i + 1
  if i == 3 :
    for i in range(0, 3):
      keyList.append(list(totalList[i].keys())[0])
      valueList.append(list(totalList[i].values())[0])
    break

print(totalList)
print(keyList)
print(valueList)
```

```python
dictionary = {'apple': '사과', 'banana': '바나나', 'cherry': '체리'}

while True:
    word = input("영어 단어를 입력하세요: ")

    # 단어가 사전에 있는 경우
    if word in dictionary:
        print(f"{word}: {dictionary[word]}")

    # 단어가 사전에 없는 경우
    else:
        print(f"{word}는(은) 사전에 없습니다.")
        answer = input("이 단어를 추가하시겠습니까? (y/n): ")

        if answer == 'y':
            meaning = input("뜻을 입력하세요: ")
            dictionary[word] = meaning
            print(f"{word}가(이) 사전에 추가되었습니다.")

        else:
            print("사전에 추가하지 않았습니다.")
    
    # 종료 여부 확인
    exit = input("종료하시겠습니까? (y/n): ")
    if exit == 'y':
        break
```

```python
import random

# 1 =  가위, 2 = 바위, 3 = 보

MAXCOUNT = 3
i = 0;
while i < MAXCOUNT :

  comResult = random.randrange(3) + 1
  userResult = 0
  userInput = input("가위,바위,보 중에 하나를 입력하세요: ")

  try:
    flag = True
    i += 1
    if userInput == "가위":
      userResult = 1
    elif userInput == "바위":
      userResult = 2
    elif userInput == "보":
      userResult = 3
    else :
      print("다시 입력하세요.")
      i -= 1
      flag = False
      
    if flag:
      if userResult - comResult == 1 or userResult - comResult == -2 :
        print("이겼습니다.")
      elif userResult == comResult :
        print("비겼습니다.")
      elif userResult - comResult == -1 or userResult - comResult == 2 :
        print("졌습니다.")
      else :
        print("오류 발생으로 승부를 알 수 없습니다.")
        i -= 1

  except Exception as e:
    print("에러 발생:", e)

    
```
