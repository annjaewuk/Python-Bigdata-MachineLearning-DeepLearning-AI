```python
left, right = 1, 1
result = ""  # 결과를 저장할 문자열 변수 선언
for right in range(1,6):
    left = 1
    for left in range(1,6):
        result += f"{left} x {right} = {left*right}\t"  # 결과를 문자열 변수에 추가
        left += 1
    result += "\n" if right != 5 else ""  # 마지막 행인 경우 개행 문자를 추가하지 않음
    right += 1
print(result)  # 결과 출력
```

```python
# 리스트, 딕션너리, 순환문, 랜덤 choice

# 어느학급에 반장선거를 한다... 
# 후보자는 다음과 같다
candidate = ["김씨","이씨","박씨","최씨","홍씨"]
# 학급학생은 20명이고  각 학생이 한명의 후보한테 투표를 한다고 하면
# 랜덤함수를 이용해서 각 후보에대해 투표를 하고   random.choice(candidate)
# 당성자를 출력
# hint  딕션너리 구조를 충분히 활용

# 투표 dictionary 만들기
vote_dic = {}
for name in candidate:
  vote_dic[name] = 0

# 순환문을 이용해서   candidate중에 한명을 선택해서 dictionary의 키로 접근한다
for i in range(20):
  vote_name = random.choice(candidate)
  vote_dic[vote_name] += 1

# 당선자이름을 저장할 변수와  최대값을 담는 변수 만들어서
# dictoinary를 순환하면서 값이 큰 값을 만날때마다 이름을 업데이트한다.
win_name, max_vote = candidate[0], vote_dic[candidate[0]]
for key,value in vote_dic.items():
  if value > max_vote:
    win_name = key

print(f"개표결과 : {vote_dic}")
print(f"당선자 : {win_name}")
```

```python
n = int(input("참가자 수를 입력하세요: "))
print("------------------------\n")

score_dict = {}

# 참가자 수만큼 이름과 점수 입력 받음
for i in range(n):
    name = input(f"{i+1}번 참가자 이름을 입력하세요: ")
    score = int(input(f"{name}의 점수를 입력하세요: "))
    print("------------------------\n")
    score_dict[name] = score

# 점수를 기준으로 내림차순 정렬
sorted_scores = sorted(score_dict.items(), key=lambda x: x[1], reverse=True)

# 등수 매기기
rank = 1
prev_score = sorted_scores[0][1]
ranked_scores = []
for i, (name, score) in enumerate(sorted_scores):
    if score != prev_score:
        rank = i+1
        prev_score = score
    ranked_scores.append((name, score, rank))

# 출력
print("┌────────┬────────┬───────┐")
print("│   순위   │   이름   │   점수   │")
print("├────────┼────────┼───────┤")
for name, score, rank in ranked_scores:
    print(f"│   {rank}위   │ {name.center(7)} │ {str(score).center(5)}점 │")
    print("├────────┼────────┼───────┤")
print("└────────┴────────┴───────┘")
```

```python
#0~100 사이의 임의 수를 랜덤으로 선택하여 저장

#사용자는 정수를 10번 입력할 기회를 갖는다.
#각 기회의 입력을 받은 후 , 그 값이 랜덤 저장된 값보다 크면, "그보다 큽니다.", 작으면 "그보다 작습니다" 를 매번 출력
#그 수를 맞추면, "축하합니다! n번 만에 맞추셨습니다", 맞추지 못하면 "그 수는 n이었습니다" 라고 출력.
import random

lottery_num = random.randint(0, 100)
MAX_TRY = 4
i = 0;

print("컴퓨터가 마음속으로 0부터 100 사이의 수를 생각했습니다.")
print("------------------------------------------------------")

while i < MAX_TRY :
  i += 1
  user_guessing = int(input(f"맞춰보세요! {i}번째 시도! 0부터 100사이의 수는?: "))
  if user_guessing == lottery_num :
    print(f"축하합니다! {i}번 만에 맞추셨습니다")
    break
  elif user_guessing < lottery_num :
    print(f"그보다 큽니다")
  elif user_guessing > lottery_num :
    print(f"그보다 작습니다")
  

print(f"그 수는 {lottery_num}이었습니다")
```
