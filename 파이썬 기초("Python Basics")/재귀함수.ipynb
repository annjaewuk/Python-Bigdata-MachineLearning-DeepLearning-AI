```python
def recursive(data): # 1번째, data = 0 / 6번째, data = 1 / 11번째, data = 2
  data += 1          # 2번째, data = 1 / 7번째, data = 2 / 12번째, data = 3
  if data == 3:      # 3번째, 넘어감   / 8번째, 넘어감 / 13번째 들어감
    return data                                        # 14번째 3을 리턴해서 10번째 리턴문으로 돌아감
  else :
    print("data:", data)          # 4번째, data: 1 출력 / 9번째 data: 2 출력
    return data + recursive(data) # 5번째, return = 1 + recursive(data), def recursive(data)로 다시 돌아감 / 10번째 return = 1 + 2 + recursive(data), def recursive(data)로 다시 돌아감 / 15번째 1+2+3 = 6

recursive(0)
```

```python
def factorial(n):
    if n == 1:          # 종료 조건: n이 1이면 1을 반환
        return 1
    else:               # 재귀 호출: n과 n-1의 팩토리얼 값의 곱을 반환
        return n * factorial(n-1)


factorial(5)
```
