```python
keep = True

def getInput():
  a = input("숫자를 입력해라: ")
  return a

while keep:
  data = getInput()
  if data.isdigit() :
    print(int(data) * 2);
    keep = False;
    data = ""
  else :
    print("숫자가 아닙니다.")
    data = ""


#이 코드에서 4를 입력했을 때, 256이 나오는 이유

#문제는 이거였음 print(int(data) ** int(data));
```

```python
#초기 변수 선언
num1 = None
num2 = None
flag = True

#입력받는 함수 선언 및 입력값 리턴(반복되므로)
def getInput():
  a1 = input("정수를 입력해라: ");
  a2 = input("정수를 입력해라: ");
  return a1, a2

#변수에 값 대입
num1, num2 = getInput()

#원하는 값을 입력받을때까지 반복
while flag:
  #둘다 정수면
  if num1.isdigit() == True and num2.isdigit() == True :
    #해당 값들을 정수로 변환
    num1 = int(num1)
    num2 = int(num2)
    #출력
    print(f"더하기는: {num1+num2}");
    print(f"빼기는: {num1-num2}");
    print(f"곱하기는: {num1*num2}")
    print(f"나누기는: {num1/num2}")
    #파이썬은 Do-While이 없으므로, 조건 성립시 Break를 통해 탈출한다.
    break
  #정수가 아닌 다른 값이면  
  else :
    print("다시하거라.")
    #다시 입력을 받는다
    num1, num2 = getInput()

#이 코드의 문제점이 있거나, 개선 사항이 있으면 알려주세요
```

```python
#개선된 코드

def get_input():
    while True:
        try:
            num1 = int(input("정수를 입력해라: "))
            num2 = int(input("정수를 입력해라: "))
            return num1, num2
        except ValueError:
            print("다시하거라.")

num1, num2 = get_input()

try:
    print(f"더하기는: {num1 + num2}")
    print(f"빼기는: {num1 - num2}")
    print(f"곱하기는: {num1 * num2}")
    print(f"나누기는: {num1 // num2}")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
    num1, num2 = get_input()
```

```python
#더욱더 개선된 코드

while True:
    num1 = input("첫 번째 정수를 입력하세요: ")
    num2 = input("두 번째 정수를 입력하세요: ")
    try:
        num1 = int(num1)
        num2 = int(num2)
        print(f"더하기: {num1 + num2}")
        print(f"빼기: {num1 - num2}")
        print(f"곱하기: {num1 * num2}")
        print(f"나누기: {num1 // num2}")
        break
    except ValueError:
        print("정수를 입력해주세요.")
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
```
