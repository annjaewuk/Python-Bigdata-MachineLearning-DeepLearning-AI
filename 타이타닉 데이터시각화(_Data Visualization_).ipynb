```python
import numpy as np
import time

def sum_array(length, method):
    if method == 'numpy':
        # 넘파이 배열 생성
        arr = np.random.rand(length)

        # 넘파이를 이용한 합 계산 및 시간 측정
        start_time = time.time()
        result = np.sum(arr)
        end_time = time.time()

        # 결과 반환
        return result, (end_time - start_time)

    elif method == 'for':
        # 리스트 생성
        lst = [i for i in range(length)]

        # 리스트를 이용한 합 계산 및 시간 측정 (for문)
        start_time = time.time()
        result = 0
        for i in lst:
            result += i
        end_time = time.time()

        # 결과 반환
        return result, (end_time - start_time)

    elif method == 'comprehension':
        # 리스트 생성
        lst = [i for i in range(length)]

        # 리스트를 이용한 합 계산 및 시간 측정 (리스트 컴프리헨션)
        start_time = time.time()
        result = sum([i for i in lst])
        end_time = time.time()

        # 결과 반환
        return result, (end_time - start_time)

    else:
        # 잘못된 메소드가 입력된 경우 예외 처리
        raise ValueError("Invalid method. Choose 'numpy', 'for', or 'comprehension'.")

# 넘파이 계산
result, elapsed_time = sum_array(100000, 'numpy')
print(f"Result: {result}")
print(f"Elapsed Time with NumPy: {elapsed_time:.5f} seconds")

# 일반적인 for문을 이용한 합 계산
result, elapsed_time = sum_array(100000, 'for')
print(f"Result: {result}")
print(f"Elapsed Time with for-loop: {elapsed_time:.5f} seconds")

# 리스트 컴프리헨션을 이용한 합 계산
result, elapsed_time = sum_array(100000, 'comprehension')
print(f"Result: {result}")
print(f"Elapsed Time with list comprehension: {elapsed_time:.5f} seconds")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
titanic = sns.load_dataset('titanic')

# 1. survived, pclass, sex, age 항목만으로 구성된 새로운 데이터 프레임
selected_columns = ['survived', 'pclass', 'sex', 'age']
new_titanic = titanic[selected_columns]

# 결측치 처리
mean_age = new_titanic['age'].mean()
new_titanic['age'].fillna(mean_age, inplace=True)

# 연령대 구하기
new_titanic['age_group'] = (new_titanic['age'] // 10) * 10

# 연령대별 사망자 수 계산
age_group_death_count = new_titanic[new_titanic['survived'] == 0].groupby('age_group')['survived'].count()

# 연령대별 승객 수 계산
age_group_total_count = new_titanic.groupby('age_group')['survived'].count()

# 연령대별 사망률 계산
age_group_death_rate = age_group_death_count / age_group_total_count * 100

# 막대 그래프로 출력
ax = age_group_death_rate.plot(kind='bar', figsize=(10, 5), ylabel='Death Rate (%)', title='Age Group Death Rate on Titanic', rot=0)
x_labels = [f"{int(age)}s" for age in age_group_death_rate.index]
ax.set_xticklabels(x_labels)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
titanic = sns.load_dataset('titanic')

# 1. survived, pclass, sex, age 항목만으로 구성된 새로운 데이터 프레임
selected_columns = ['survived', 'pclass', 'sex', 'age']
new_titanic = titanic[selected_columns]

# 결측치 처리
mean_age = new_titanic['age'].mean()
new_titanic['age'].fillna(mean_age, inplace=True)

# 연령대 구하기
new_titanic['age_group'] = (new_titanic['age'] // 10) * 10

# 연령대별, 성별, pclass별 생존자 수 계산
grouped_survival_count = new_titanic.groupby(['age_group', 'sex', 'pclass', 'survived']).size().unstack(fill_value=0)

# 연령대별, 성별, pclass별 생존률 계산
grouped_survival_rate = (grouped_survival_count[1] / grouped_survival_count.sum(axis=1)) * 100
grouped_survival_rate = grouped_survival_rate.reset_index()

# 히트맵 데이터 구성
heatmap_data = grouped_survival_rate.pivot_table(values=0, index=['sex', 'pclass'], columns='age_group', fill_value=0)

# 히트맵 그리기
plt.figure(figsize=(12, 5))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt='.1f', cbar_kws={'label': 'Survival Rate (%)'})
plt.title('Survival Rate by Age Group, Gender, and Pclass')
plt.xlabel('Age Group')
plt.ylabel('Gender and Pclass')
plt.show()
```
