범주형 or 문자형 데이터 처리

```python
# object 타입을 찾아서
df.dtypes[df.dtypes.values == 'O'].index
```

```python
# category 타입으로 변경하기
for col in df.dtypes[df.dtypes.values == 'O'].index:
  df[col] = df[col].astype('category')
df.info()  
```

```python
# 라벨인코딩(이진형태), 원핫(다중형태)
pd.get_dummies(df)
```

```python
df_onehot = pd.get_dummies(df)
print(df.shape, df_onehot.shape)
df_onehot.info()
```

```python
df_onehot.isnull().mean()
```

결측치 처리 Tip
  - 대체
    - 평균,중위수,최소,최대,표준편차 : 수치형 데이터
    - 가장많은수의 문자열 : 문자열
  - 제거
    - 극히 적거나(결측치의 비율), 대처할 다른 수단이 없을때

Newbie별 평균값으로나이를 대체

```python
df_onehot['Age'] = df_onehot['Age'].fillna(df_onehot.groupby('Newbie')['Age'].transform(lambda x: x.mean()))
# df_onehot.groupby('Newbie')['Age'].transform(lambda x: x.mean())
df_onehot.info()
```

```python
# 모델적용
# 성능비교.... 스케일링 여부
# 1. onehot으로 변경만 하고 연속형 수치데이터는 Age 밖에 없으니 스케일링 안함
# 2. Age 만 스케일링
# 3. 전체 데이터셋 스케일링
```

```python
X = df_onehot.iloc[:,1:]
y = df_onehot.iloc[:,0]
```

```python
# 교차검증
import numpy as np
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
score = cross_validate(LogisticRegression(),X,y)
np.mean(score['test_score'])
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
df_onehot_copy = df_onehot.copy(deep=True)
df_onehot_copy['Age'] = StandardScaler().fit_transform(df_onehot_copy[['Age']])
```

```python
X = df_onehot_copy.iloc[:,1:]
y = df_onehot_copy.iloc[:,0]
score = cross_validate(LogisticRegression(),X,y)
np.mean(score['test_score'])
```

```python
df_onehot_copy = df_onehot.copy(deep=True)
X = df_onehot_copy.iloc[:,1:]
y = df_onehot_copy.iloc[:,0]
X = StandardScaler().fit_transform(X)
score = cross_validate(LogisticRegression(),X,y)
np.mean(score['test_score'])
```

```python
lr = LogisticRegression()
x_train,x_test,y_train,y_test= train_test_split(X,y,random_state=10)
```
