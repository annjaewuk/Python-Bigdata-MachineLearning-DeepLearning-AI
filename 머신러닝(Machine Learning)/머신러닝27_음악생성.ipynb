```python
import json
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

매개변수에 따라서 톤을 합성

```python
def tone_make(freq, duration, amplitude=1.0, sampling_freq = 44100):
  # print(freq, duration, amplitude,sampling_freq)
  # 시각 축 생성    
  time_axis = np.linspace(0,duration, duration*sampling_freq)
  # 오디오 신호 생성    
  signal = amplitude*np.sin(2*np.pi*freq*time_axis)  
  return signal.astype(np.int16) 
```

```python
# 오디오 출력파일 이름
file_tone_single = 'generated_single.wav'
file_tone_sequence = 'generated_sequence.wav'
mapping_file = '/content/mapping_27.json'
with open(mapping_file, 'r') as f:
  tone_map =  json.loads(f.read())
print(tone_map)  
```

```python
# 3초인 F톤
tone_name = 'F'
duration = 3
#진폭
amplitude = 12000
# 주파수
sampling_freq = 44100
```

```python
# 대응하는 톤 주파수
tone_freq =  tone_map[tone_name]
```

```python
# 톤을 생성
make_tone = tone_make(tone_freq,duration,amplitude,sampling_freq)
```

```python
# 오디오 신호를 출력파일으로 생성
write(file_tone_single,sampling_freq,make_tone)
```

```python
# 음악처럼 만들기위해 톤 순서를 정함
tone_sequence = [('G',0.4),('D',0.5),('F',0.3),('C',0.6),('A',0.4)]
for item in tone_sequence:
  print(item)
```

```python
# 톤순서로 오디오 신호를 생성
signal = np.array([])
for item in tone_sequence:
  #톤 이름 가져오기
  tone_name =  item[0]
  # 각 톤에 대응하는 주파수를 추출
  freq = tone_map[tone_name]
  # 지속시간
  duration = item[1]
  # 톤 합성  
  synthesized_tone = tone_make(freq,duration,amplitude,sampling_freq)
  # 출력신호에 추가
  signal = np.append(signal,synthesized_tone, axis=0 )
# 저장
write(file_tone_sequence,sampling_freq,signal)
```
