```python
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from lightgbm import LGBMClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

scaling

```python
best_score = 0
for scaler in [None, MinMaxScaler(), StandardScaler()]:
  X = pd.get_dummies(X.copy())
  test_df = pd.get_dummies(test_df.copy())
  if scaler == None:
    train_x_scaled = X.copy()
    text_x_scaled = test_df.copy()
  else:
    scaler.fit(X)
    train_x_scaled = scaler.transform( X.copy())
    text_x_scaled = scaler.transform(test_df.copy())


  skf = StratifiedKFold(n_splits=5,shuffle=True)
  model = LGBMClassifier(n_jobs=-1)
  scores = cross_val_score(model,train_x_scaled,y,scoring='accuracy',cv=skf,n_jobs=1)
  print(f'최대성능 : {max(scores)} 평균: {np.mean(scores)}')
  print(f'-'*100)
  if np.mean(scores) >= best_score:
      best_score = np.mean(scores)
      best_train_x = train_x_scaled
      best_test_x = text_x_scaled
      best_scaler = scaler
if best_scaler == None:
    print('BEST SCALER IS NON_SCALED')
else:
    print('BEST SCALER : ', best_scaler.__class__.__name__)
```
