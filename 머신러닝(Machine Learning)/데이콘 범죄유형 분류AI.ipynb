원핫인코딩
  - 순서가 없는 범주형 데이터로서 순서가 중요하지 않고 독립적인 특성

```python
X.info()
```

```python
# 인코딩은 원핫
# 월, 요일, 범죄발생지
```

```python
X.describe()
```

```python
# 원핫으로 변경하기
def makeOneHot(colname):
  onehot_month = ohe.fit_transform( X[colname].to_numpy().reshape(-1,1) ).toarray()  
  return pd.DataFrame(onehot_month,columns= [ str(i) for i in ohe.categories_[0] ])

onehot_df = pd.concat([  makeOneHot('월'),makeOneHot('요일'),makeOneHot('범죄발생지') ],axis=1 )
onehot_df.shape
```

```python
temp_x = X.loc[:,X.columns[3:10]]
# 표준화
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(temp_x)
temp_x_scaled = ss.transform(temp_x)
new_x = pd.concat([ pd.DataFrame(temp_x_scaled,columns=temp_x.columns), onehot_df],axis=1)
new_x.shape
```

```python
X.shape,new_x.shape, y.shape
```

```python
rfc = RandomForestClassifier()
from sklearn.model_selection import cross_validate
scores = cross_validate(rfc,new_x, y, return_train_score=True)
scores
```

```python
params = {'max_depth' : [None,2,3,4,5,6,7,8,9,10,20]}
from sklearn.model_selection import GridSearchCV
grid_reg = GridSearchCV(RandomForestClassifier(),param_grid=params)
grid_reg.fit(new_x,y)
# 최상의 매개변수 출력
print(f'최상의 매개변수 : {grid_reg.best_params_}  최상의 점수 : {grid_reg.best_score_}')
```

```python
test_X = test_df.iloc[:,1:]
test_X.shape
def makeOneHotTest(colname):
  ohe = OneHotEncoder()
  onehot_month = ohe.fit_transform( test_X[colname].to_numpy().reshape(-1,1) ).toarray()  
  return pd.DataFrame(onehot_month,columns= [ str(i) for i in ohe.categories_[0] ])

onehot_df = pd.concat([  makeOneHotTest('월'),makeOneHotTest('요일'),makeOneHotTest('범죄발생지') ],axis=1 )

temp_test_x = test_X.loc[:,test_X.columns[3:10]]  # 이진형태의 데이터 제외하고 연속적인 수치형 데이터만 추출
# 표준화
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(temp_test_x)
temp_test_x_scaled = ss.transform(temp_test_x)

new_test_x = pd.concat([ pd.DataFrame(temp_test_x_scaled,columns=temp_test_x.columns), onehot_df],axis=1)
new_test_x.shape
```
